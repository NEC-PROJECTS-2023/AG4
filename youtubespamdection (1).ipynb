```python
import numpy as np
import pandas as pd
import seaborn as sns

import missingno as ms
%matplotlib inline
train_data = []
data_files = ['/content/drive/MyDrive/youtube/Youtube01.csv','/content/drive/MyDrive/youtube/Youtube02-KatyPerry.csv','/content/drive/MyDrive/youtube/Youtube03-LMFAO.csv','/content/drive/MyDrive/youtube/Youtube04-Eminem.csv','/content/drive/MyDrive/youtube/Youtube05-Shakira.csv']
for file in data_files:
    data = pd.read_csv(file)
    train_data.append(data)
train_data = pd.concat(train_data)
train_data.head(15)

## Function which drops the given features from the given dataframe
```

```python
def drop_fectures(features,data):
    data.drop(features,axis=1,inplace=True)
drop_fectures(['COMMENT_ID','AUTHOR','DATE'],train_data)
train_data.info()
```

```python
import re
re.findall("[A-Za-z]+","Hey Ravi57,How are you doing?")
```

```python
def process_content(content):
    return " ".join(re.findall("[A-Za-z]+",content.lower()))
```

```python
train_data['processed_content'] = train_data['CONTENT'].apply(process_content)
```

```python
train_data.head(15)
```

```python
drop_fectures(['CONTENT'],train_data)
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_data['processed_content'],train_data['CLASS'],test_size=0.2,random_state=57)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(stop_words='english')
x_train_counts = count_vect.fit_transform(x_train)
x_train_counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tranformer = TfidfTransformer()
x_train_tfidf = tranformer.fit_transform(x_train_counts)
x_train_tfidf.shape
```

```python
x_test_counts = count_vect.transform(x_test)
x_test_counts
```

```python
x_test_tfidf = tranformer.transform(x_test_counts)
x_test_tfidf
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_tfidf,y_train)
```

```python
predictions = model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import matthews_corrcoef 
print(matthews_corrcoef(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train_tfidf,y_train)
```

```python
predictions = model.predict(x_test_tfidf)
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import matthews_corrcoef 
print(matthews_corrcoef(y_test,predictions))
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model = BernoulliNB()
model.fit(x_train_tfidf,y_train)
```

```python
predictions = model.predict(x_test_tfidf)
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import matthews_corrcoef 
print(matthews_corrcoef(y_test,predictions))
```
